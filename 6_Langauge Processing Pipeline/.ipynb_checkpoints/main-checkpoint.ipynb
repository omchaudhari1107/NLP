#### thus spacy provide pretrained pipline

how to install it for english, but for more visit : https://spacy.io/models
```
python -m spacy download en_core_web_sm
```

````python
nlp = spacy.load('en_core_web_sm')
````

````python
nlp.pipe_names
````

````python
doc = nlp("Let's go to N.Y.!")
for token in doc:
    print(token," | ",token.pos_," | ",token.lemma_)
````

````python
doc = nlp('Tesla is worth of 578.94 billion USD')

for ent in doc.ents:
    print(ent.text," | ",ent.label_," | ",spacy.explain(ent.label_))
````

````python
from spacy import displacy

displacy.render(doc,style='ent')
````

### if i only want to use 'ner' from the pipeline so what did i do?

````python
src_nlp = spacy.load('en_core_web_sm')
nlp = spacy.blank('en')
nlp.add_pipe('ner',source=src_nlp)
````

````python
nlp.pipe_names
````

````python
doc = nlp('Tesla is worth of 578.94 billion USD')

for ent in doc.ents:
    print(ent.text," | ",ent.label_," | ",spacy.explain(ent.label_))
````
